```python
from bayes_opt import BayesianOptimization
from modeling.bert_base_multilingual.cased.hyperparameters.optimization_structure import evaluate_model
from modeling.baseline_models.hyperparameters.plots.gaussian_process_plot import plot_gp

# For saving progress
# from bayes_opt.logger import JSONLogger
# from bayes_opt.event import Events

# For loading progress
from bayes_opt.util import load_logs
```

# Load logs

```python
# Bounded region of parameter space
pbounds = {
    'max_token_count': (33, 60),
    'epochs': (1, 1.1),
    'batch_size': (8, 128),
    'learning_rate': (1e-4, 5e-5),
    'dropout': (0, 1),
    'learning_rate_schedule': (0, 1),
}

optimizer = BayesianOptimization(
    f=evaluate_model,
    pbounds=pbounds,
    verbose=2,  # verbose=1 prints only when a max is observed
    random_state=1,
)
```

```python
load_logs(optimizer, logs=["hyperparameter_optimization/bert_base_multilingual_cased/logs/logs.json"])
```
